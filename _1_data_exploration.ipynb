```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_bio = pd.read_csv('original_datasets/biometrics_data_combined.csv')
df_demo = pd.read_csv('original_datasets/demographics_data_combined.csv')
df_enrol = pd.read_csv('original_datasets/enrolment_data_combined.csv')
```

```python
def explore_dataset(df, category_name):
    print(f"\n{'='*50}")
    print(f"Dataset Name: {category_name}")
    print(f"{'='*50}")
    print(f"Shape: {df.shape}")
    print(f"\nTotalColumns: {df.columns.tolist()}")
    print(f"\nData Types:\n{df.dtypes}")
    print(f"\nMissing Values:\n{df.isnull().sum()}")
    print(f"\nFirst 5 rows:\n{df.head()}")
    print(f"\nNumeric Summary:\n{df.describe()}")
```

```python
explore_dataset(df_bio, "Biometrics Data")
explore_dataset(df_demo, "Demographics Data")
explore_dataset(df_enrol, "Enrolment Data")
```

```python
#sort according to pincodes
df_bio = df_bio.sort_values(by='pincode')
df_demo = df_demo.sort_values(by='pincode')
df_enrol = df_enrol.sort_values(by='pincode')
```

```python
# Get the first district name for each 3-digit pincode prefix
district_mapping = df_bio.groupby(df_bio['pincode'].astype(str).str[:3])['district'].first().to_dict()

# Apply the mapping efficiently
pincode_prefix = df_bio['pincode'].astype(str).str[:3]
df_bio['district'] = pincode_prefix.map(district_mapping)

# Apply to other dataframes
pincode_prefix_demo = df_demo['pincode'].astype(str).str[:3]
df_demo['district'] = pincode_prefix_demo.map(district_mapping)

pincode_prefix_enrol = df_enrol['pincode'].astype(str).str[:3]
df_enrol['district'] = pincode_prefix_enrol.map(district_mapping)
```
